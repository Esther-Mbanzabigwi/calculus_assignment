# Task 1: Write a dynamic function to find the derivative of any function f(x)

```python
import sympy as sp

def dy_dx(f):
    # Define the symbol
    x = sp.symbols('x')

    # Calculate the derivative
    derivative = sp.diff(f, x)

    return derivative

# Example usage
# Define the function f(x)
x = sp.symbols('x')
f = x**2 + 3*x + 2

# Calculate the derivative
derivative = dy_dx(f)
print(derivative)
```

#Task 3: Plot a graph of the Equation must have 4 points (2 minimas and 2 maximas)

import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 2*x**4 - 8*x**3 - 6*x**2 + 4*x + 5

# Generate x values
x = np.linspace(-2, 3, 400)
# Calculate corresponding y values
y = f(x)

# Find minima and maxima
from scipy.signal import argrelextrema

# Find local minima
minima_indices = argrelextrema(y, np.less)[0]
maxima_indices = argrelextrema(y, np.greater)[0]

minima_x = x[minima_indices]
minima_y = y[minima_indices]

maxima_x = x[maxima_indices]
maxima_y = y[maxima_indices]

# Plot the function
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x) = 2x^4 - 8x^3 - 6x^2 + 4x + 5', color='b')
plt.scatter(minima_x, minima_y, color='red', label='Minima')
plt.scatter(maxima_x, maxima_y, color='green', label='Maxima')

plt.title('Graph of f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)  # horizontal line at y=0

# Adding annotations for minima
for i in range(len(minima_x)):
    plt.text(minima_x[i], minima_y[i], f'({minima_x[i]:.2f}, {minima_y[i]:.2f})', fontsize=10, verticalalignment='bottom')

# Adding annotations for maxima
for i in range(len(maxima_x)):
    plt.text(maxima_x[i], maxima_y[i], f'({maxima_x[i]:.2f}, {maxima_y[i]:.2f})', fontsize=10, verticalalignment='top')

plt.legend()
plt.show()




# Task 4: Create separate array of Minimas and maximas

```python

import numpy as np
from scipy.optimize import fsolve

# Define the modified function
def f(x):
    return x**5 - 5*x**3 + 10*x

# Define the derivative of the function
def df(x):
    return 5*x**4 - 15*x**2 + 10

# Generate x values
x = np.linspace(-3, 3, 400)

# Generate y values for the function and its derivative
y = f(x)
dy = df(x)

# Initial guesses for the roots (critical points)
initial_guesses = [-2, -1, 0, 1, 2]
critical_points = fsolve(df, initial_guesses)

# Calculate function values at critical points
critical_values = f(critical_points)

# Filter out the critical points that are minima and maxima
minima = [cp for cp, val in zip(critical_points, critical_values) if df(cp + 1e-5) > 0]
maxima = [cp for cp, val in zip(critical_points, critical_values) if df(cp + 1e-5) < 0]

# Create arrays to store minima and maxima
arr_minimas = list(zip(minima, f(np.array(minima))))
arr_maximas = list(zip(maxima, f(np.array(maxima))))

# Print the arrays of minima and maxima
print(f'The array of minimas is {arr_minimas} and the array of maximas is {arr_maximas}')




```
